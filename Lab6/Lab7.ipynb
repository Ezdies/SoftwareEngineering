```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

x = np.linspace(-np.pi/2, np.pi/2, 100)
y = np.linspace(-np.pi/2, np.pi/2, 100)

X, Y = np.meshgrid(x, y)

Z = np.sin(X*2) + np.cos(3*Y) + np.sin(X*Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(X, Y, Z, cmap="turbo")

def update(frame):
    ax.view_init(elev=30, azim=frame)
    return surf,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 360, 36), interval=50, blit=True)

plt.show()
```

```python
def julia(z, c):
    for i in range(100):
        z = z*z + c
        if abs(z) > 2: return i
    
    return 0

c = 0.8 + 0.156j
X, Y = np.meshgrid(np.linspace(-1.5, 1.5, 500), np.linspace(-1.5, 1.5, 500))
Z = X + Y * 1j

color = np.frompyfunc(julia, 2, 1)(Z, c).astype(np.float64)


plt.figure(figsize=(10, 10))
plt.contourf(X, Y, color, levels = np.linspace(0, 100, 100), cmap='jet')
plt.show()



```
